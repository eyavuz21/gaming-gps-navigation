```python
#using mahalanobis distance to remove outliers 

import pandas as pd 

correcteddf = pd.read_csv(r'/Users/macbook/Desktop/UCL PhD Work/SHQoutputs/PREPROCESSING/correcteddfmalefemale'+'.csv')

import numpy as np 

correcteddf = correcteddf.rename(columns={"response-1 -quantised":'response1q',"response-2-quantised":'response2q',"response-3-quantised":'response3q',"response-4-quantised":'response4q',"response-5-quantised":'response5q',"response-6-quantised":'response6q',"response-7-quantised":'response7q',"response-8-quantised":'response8q',"response-9-quantised":'response9q',"response-10-quantised":'response10q',"response-11-quantised":'response11q',"response-12-quantised":'response12q',"response-13-quantised":'response13q',"response-14-quantised":'response14q',"gps1-quantised":'gps1q',"gps2-quantised":'gps2q',"gps3-quantised":'gps3q',"gps4-quantised":'gps4q',"gps6-quantised":'gps6q',"gps7-quantised":'gps7q'})

#recoding the NSQ subscale responses

correcteddf['response1q'][correcteddf['response1q'] == 3] = 1.5
correcteddf['response6q'][correcteddf['response6q'] == 3] = 1.5
correcteddf['response7q'][correcteddf['response7q'] == 3] = 1.5
correcteddf['response14q'][correcteddf['response14q'] == 3] = 1.5

#calculating the average NSQ ability and NSQ strategy score

correcteddf['NSQaverage'] = correcteddf[['response1q','response2q','response3q','response4q','response5q','response6q','response7q','response8q','response9q','response10q','response11q','response12q','response13q','response14q']].mean(axis=1)
correcteddf['NSQabilityaverage'] = correcteddf[['response2q','response3q','response4q','response12q','response13q']].mean(axis=1)
correcteddf['NSQstrategyaverage'] = correcteddf[['response6q','response5q','response11q']].mean(axis=1)
GPScolumns = ['gps1q','gps2q','gps3q','gps4q','gps6q','gps7q']
correcteddf['GPSaverage'] = correcteddf[GPScolumns].mean(axis=1)

from scipy.stats import chi2
  
def calculateMahalanobis(y=None, data=None, cov=None):
  
    y_mu = y - np.mean(data)
    if not cov:
        cov = np.cov(data.values.T)
    inv_covmat = np.linalg.inv(cov)
    left = np.dot(y_mu, inv_covmat)
    mahal = np.dot(left, y_mu.T)
    return mahal.diagonal()

correcteddf['Mahalanobis'] = calculateMahalanobis(y=correcteddf[['zscore','NSQaverage','GPSaverage','age','video_game_all_devices_hours_per_week','hours_of_phone_use_per_week']],data=correcteddf[['zscore','NSQaverage','GPSaverage','age','video_game_all_devices_hours_per_week','hours_of_phone_use_per_week']])
correcteddf['m_p'] = 1 - chi2.cdf(correcteddf['Mahalanobis'], 6)
```

```python
#using winsorised method to remove outliers (replace >3SD above mean with 3SD above mean)

import numpy as np 
import pandas as pd 

correcteddf2 = pd.read_csv('/Users/macbook/Desktop/UCL PhD Work/SHQoutputs/PREPROCESSING/NEWSAMPLEMARYFORPLOTS.csv')
columns_to_process = [
    'zscore',
    'age', 
    'GPSaverage', 
    'hours_of_phone_use_per_week', 
    'video_game_all_devices_hours_per_week'
]
for col in columns_to_process:
    mean_val = correcteddf2[col].mean()
    std_val = correcteddf2[col].std()
    lower_limit = mean_val - 3 * std_val
    upper_limit = mean_val + 3 * std_val
    correcteddf2[col] = np.where(correcteddf2[col] < lower_limit, lower_limit, correcteddf2[col])
    correcteddf2[col] = np.where(correcteddf2[col] > upper_limit, upper_limit, correcteddf2[col])

correcteddf2.to_csv(r'/Users/macbook/Desktop/UCL PhD Work/SHQoutputs/PREPROCESSING/mostrecentnewsampleMary3SD'+'.csv')
```

```python
#using IQR (more conservative) method to remove outliers

correcteddf2 = pd.read_csv('/Users/macbook/Desktop/UCL PhD Work/SHQoutputs/PREPROCESSING/NEWSAMPLEMARYFORPLOTS.csv')
columns_to_process = [
    'age', 
    'GPSaverage', 
    'hours_of_phone_use_per_week', 
    'video_game_all_devices_hours_per_week'
]
for col in columns_to_process:
    q1 = correcteddf2[col].quantile(0.25)
    q3 = correcteddf2[col].quantile(0.75)
    iqr = q3 - q1
    lower_limit = q1 - 1.5 * iqr
    upper_limit = q3 + 1.5 * iqr
    correcteddf2[col] = np.where(correcteddf2[col] < lower_limit, lower_limit, correcteddf2[col])
    correcteddf2[col] = np.where(correcteddf2[col] > upper_limit, upper_limit, correcteddf2[col])
correcteddf2.to_csv('/Users/macbook/Desktop/UCL PhD Work/SHQoutputs/PREPROCESSING/mostrecentnewsampleMaryIQR.csv', index=False)
```
